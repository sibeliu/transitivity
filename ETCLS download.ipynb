```python
#load html files from ETCSL, clean them, save them to disk
#RUn this before EADH v.4

from pathlib import Path
import os
import re
import pandas as pd
import numpy as np

path = './mesopotamia/'

## read in the .html files from https://etcsl.orinst.ox.ac.uk/index1.htm for sumerian texts
from urllib.request import urlopen
from bs4 import BeautifulSoup

url_list = ["https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.1.1.1", 
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.1.1.2", 
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.1.1.3",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.1.1.4", 
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.1.2.1",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.1.2.2", 
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.1.3.1",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.1.3.2",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.1.3.3",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.1.3.4",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.1.3.5",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.1.4.1.1",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.1.4.1.3",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.1.4.1",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.1.4.3",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.1.4.4",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.1.5.1",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.1.6.1",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.1.6.2",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.1.6.3",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.1.7.1", 
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.1.7.3", 
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.1.7.4",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.1.7.6",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.1.7.7",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.1.7.8",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.1.8.1.1",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.1.8.1.2",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.1.8.1.3",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.1.8.1.4",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.1.8.1.5.1",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.1.8.1.5",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.1.8.2.1",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.1.8.2.2",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.1.8.2.3",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.1.8.2.4",
            
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.01.1",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.02.1",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.03.1",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.05.1",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.06.1",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.07.1",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.07.2",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.07.3",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.07.4",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.07.5",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.07.6",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.07.7",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.07.8",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.07.9",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.07.a",
            
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.08.1",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.08.2",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.08.3",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.08.4",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.08.5",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.08.6",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.08.7",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.08.8",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.08.9",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.08.10",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.08.12",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.08.13",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.08.15",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.08.16",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.08.18",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.08.20",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.08.22",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.08.23",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.08.25",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.08.26",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.08.28",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.08.29",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.08.30",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.08.31",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.08.32",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.08.33",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.08.a",
            
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.12.1",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.12.2",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.13.01",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.13.02",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.13.03",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.13.04",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.13.05",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.13.06",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.13.07",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.13.08",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.13.09",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.13.10",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.13.11",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.13.12",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.13.13",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.13.14",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.13.15",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.13.a",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.13.b",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.13.c",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.13.d",
            
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.14.1",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.14.2",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.14.3",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.15.2",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.15.3",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.16.1",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.16.2",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.17.1",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.19.1",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.19.2",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.19.3",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.19.4",
            
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.21.1",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.22.1",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.22.2",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.22.3",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.22.4",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.22.5",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.22.6",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.23.1",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.24.1",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.25.1",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.25.2",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.26.1",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.27.01",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.27.02",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.27.03",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.27.04",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.27.06",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.27.07",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.27.a",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.28.1",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.29.1",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.29.2",
            
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.30.1",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.31.1",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.32.2",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.32.e",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.32.f",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.33.1",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.33.2",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.80.1",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.80.2",
            "https://etcsl.orinst.ox.ac.uk/cgi-bin/etcsl.cgi?text=t.4.80.4",
            
            
            
            
            
            
            
            
           ]

text_list = []
cleanname =[]

for url in url_list:
    html = urlopen(url).read()
    soup = BeautifulSoup(html, features="html.parser")

    # kill all script and style elements
    for script in soup(["script", "style"]):
        script.extract()    # rip it out

    # get text
    text = soup.get_text()

    # break into lines and remove leading and trailing space on each
    lines = (line.strip() for line in text.splitlines())
    # break multi-headlines into a line each
    chunks = (phrase.strip() for line in lines for phrase in line.split("  "))
    # make one long string
    text = ' '.join(chunk for chunk in chunks if chunk)
    # get rid of (parenthetical comments)
    text = re.sub("[\(\[].*?[\)\]]", "", text)
    # add newlines to replace any numbers in the text
    text = re.sub('\d', '\n', text)
    # break into sentences
    text = re.sub("\. ", ".\n", text)
    #get rid of excess newlines
    text = re.sub("\n\n", "\n", text)
    text = re.sub("\n\n", "\n", text)
    text = re.sub("\n\n", "\n", text)

    #get rid of header and footer 
    try:
        head_str = "translation"
        text = text[text.index(head_str):]

        # get rid of footer
        foot_str = "Revision history"

        # slicing off after length computation
        text = text[:text.index(foot_str)]
        text_list.append(text)

        from urllib.parse import urlparse
        filename = urlparse(url)
        name = filename[4]
        cleanname = name[7:]   #this is the name of the new files to write
        
    except Exception:  #sometimes one of the internal conditions is not met
        pass 
    
    with open(path + cleanname + '.txt', 'w') as f:
        f.write(text)
       

```
